```python
"""
Implementing rock-Spock-paper-lizard-scissors using basics of python.
Dated: 21-04-2020
Aditya Raj
"""

def name_to_number(name):
    """
    Take string name as input and returns corresponding integer.
    rock-1, spock-2, paper-3, lizard-4, scissors-5.
    """
    if (name=="rock"):
        return 0
    elif (name=="Spock"):
        return 1
    elif (name=="paper"):
        return 2
    elif (name=="lizard"):
        return 3
    elif (name=="scissors"):
        return 4
    else:
        return ("Invalid Input")

    
def number_to_name(number):
    """
     Take integer as input and returns corresponding string name.
    1-rock, 2-spock, 3-paper, 4-lizard, 5-scissors.
    """
    if (number==0):
        return("rock")
    elif (number==1):
        return("Spock")
    elif (number==2):
        return("paper")
    elif (number==3):
        return("lizard")
    elif (number==4):
        return("scissors")
    else:
        return("Invalid Input")
from random import randint
```

```python
print('''Let's Play "rock-Spock-paper-lizard-scissors"''')
player=input("Enter your name: ")
print("Hii",player)
print("Enter choice: \n0-rock \n1-Spock \n2-paper \n3-lizard \n4-scissors")
comp_point=0
user_point=0
while True:
    user_num=int(input(""+player+"'s Option: "))
    while user_num>4 or user_num<0:
        user_num=int(input("Enter valid Input: "))
    user_name=number_to_name(user_num)
    #using randint function to generate random integers.
    comp_num=randint(0,4)
    comp_name=number_to_name(comp_num)
    print(""+player+"'s Choice: ",user_name)
    print("Computer's Choice: ",comp_name)
    #logic for satisfying the criteria of game.
    difference=(comp_num-user_num)%5
    if (0<difference<=2):
        print("Computer Wins as ('",comp_name,"> ",user_name,"')")
        comp_point=comp_point+1
    elif (difference==0):
        print("Tie")
    else:
        print(player,"wins as ('",user_name,">",comp_name,"')")
        user_point=user_point+1
    print("Want to continue: (Y/N)")
    ans=input()
    if ans=='n' or ans=='N':
        print(""+player+"'s point=",user_point)
        print("Computer's point=",comp_point)
        break
    
    
```
